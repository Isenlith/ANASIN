```
import numpy as np
from matplotlib import pyplot as plt

N1 = -6
N2 = 6
N = N2 - N1 + 1

n = np.linspace(N1, N2, N)
x = np.power(n, 2)

plt.stem(n, x)
print(n)
```

```
import numpy as np
from matplotlib import pyplot as plt

N1_A = 0
N2_A = 10
NA = N2_A - N1_A + 1
n_A = np.linspace(N1_A, N2_A, NA)

N1_B = -10
N2_B = -1
NB = N2_B - N1_B + 1
n_B = np.linspace(N1_B, N2_B, NB)

x_A = 1/(2 + n_A)
x_B = 1/(3 * n_B)

n = n_A.tolist() + n_B.tolist()
x = x_A.tolist() + x_B.tolist()

plt.stem(n, x)
```
