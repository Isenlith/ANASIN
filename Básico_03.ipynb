##Decomposi√ß√£o de Sinais em Par-√≠mpar

```
plt.rcParams['figure.figsize'] = [18,5]

t = np.linspace(-10,10,100)

x = np.sin(t) + np.cos(t)
xpar = (x + x[::-1])/2
ximpar = (x - x[::-1])/2

plt.subplot(1,3,1)
plt.plot(t,xpar,'r')
plt.axvline(0,0,color ='k')
plt.axhline(0,0,color ='k')

plt.subplot(1,3,2)
plt.plot(t,ximpar,'g')
plt.axvline(0,0,color ='k')
plt.axhline(0,0,color ='k')

plt.subplot(1,3,3)
plt.plot(t,x)
plt.axvline(0,0,color ='k')
plt.axhline(0,0,color ='k')
```
