```
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [10,5]
```

$x(t) = \cos(t)e^{-0.1t}$, $\{t \in \mathcal{R} : 0 \leq t \leq 20  \}$

```
t = np.linspace(0,20,100)
n = np.cos(t)*np.exp(-0.1*t)

plt.grid()
plt.plot(t,n)
```

```
n = np.linspace(0,40,41)
t = np.linspace(0,20,100)

x_cont = np.cos(t) * np.exp(-0.1*t)

fs = 2
Ts = 1/fs

x_disc = np.cos(n*Ts)*np.exp(-0.1*n*Ts)

plt.grid()
plt.stem(n/fs,x_disc)
plt.plot(t,x_cont)
```

##Sinais Não Periódicos

```
t = np.linspace(0,1*(2*np.pi),1000)
f0 = 5 # 5 Hz

x1 = np.sin((f0*2*np.pi)*t)

plt.grid()
plt.plot(t,x1)
```

```
plt.rcParams['figure.figsize'] = [14,5]

t = np.linspace(0,5 * 2 * np.pi, 1000)

b = 2 * np.pi

x1 = np.sin(t)
x2 = np.sin(t + b)


plt.subplot(1,2,1)
plt.grid()
plt.plot(t,x1)
plt.subplot(1,2,2)
plt.grid()
plt.plot(t,x2)
```

```
plt.rcParams['figure.figsize'] = [14,5]

t = np.linspace(0,5 * 2 * np.pi, 1000)

b = 2 * np.pi

x1 = np.sin(t) * np.exp(-0.1*t)
x2 = np.sin(t + b) * np.exp(-0.1*(t + b))


plt.subplot(1,2,1)
plt.grid()
plt.plot(t,x1)
plt.subplot(1,2,2)
plt.grid()
plt.plot(t,x2)
```
