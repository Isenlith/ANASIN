```
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

```
N = 1 # Tempo de amostragem
t = np.linspace(0,N,1000)

f01 = 4 # Frequência do primeiro sinal
f02 = 12 # Frequência do segundo sinal

x1 = np.sin(f01*2*np.pi*t)
x2 = np.sin(f02*2*np.pi*t)
x = x1 + x2

plt.plot(t,x)
```

```
N = 1 # Tempo de amostragem
t = np.linspace(0,N,1000)

f0 = 2 # Frequência do sinal
A = 5

x = A*np.sin(f0*2*np.pi*t)


plt.plot(t,x)
```

```
f0 = 4
t = np.linspace(0,1,1000,endpoint=False)
x_quadrado = signal.square(2*np.pi*f0*t)

plt.plot(t,x_quadrado)
```

```
N = 5 # Tempo de amostragem
t = np.linspace(0,N,1000)
f01 = 1.3
f02 = 1.5
f03 = 1.8

x1 = np.sin(2 * np.pi * f01*t)
x2 = np.sin(2 * np.pi * f02*t)
x3 = np.sin(2 * np.pi * f03*t)

x = x1 + x2 + x3

plt.plot(t,x)
```

```
x = np.random.randn(5000,5000)

hist, bins = np.histogram(x,100)

plt.bar(bins[1::],hist)
```

```
f0 = 50
t = np.linspace(0,1,1000)

x = np.sin(2 * np.pi * f0 * t)
n = np.random.randn(t.shape[0])
x_noise = x * n

plt.plot(t,x)
plt.plot(t,x_noise,'r')
```
