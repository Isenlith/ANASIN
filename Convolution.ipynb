```
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [24, 5]
```

```
x = [4, 3, 2, 1]
h = [2, 2, -1]


def conv(a,b,n):
  if n == 0:
    t = a[0]*b[0]
  else:
    t = 0
    for i in range(0, n):
      t = t + a[i]*b[n - i]
        
  return t

y = []
y.append(conv(x, h, 0))
y.append(conv(x, h, 1))
y.append(conv(x, h, 2))
print(y)
```

```
h = [4, 3, 2, 1, 3]
x = [2, 2, -1, 2]
y = np.convolve(x, h)

print(len(h))
print(len(x))
print(len(y))

print(y)
```

```
n = np.linspace(0, 50, 51)

mu = 0
sigma = 0.1
N = np.random.randn(n.shape[0]) * sigma + mu

y = np.sin(2*np.pi*(1/50)*n) + N
k =  np.ones(40)

y_conv_k = np.convolve(y, k)

plt.subplot(1,2,1)
plt.stem(n, y)
plt.subplot(1,2,2)
plt.stem(y_conv_k)
```

```
n = np.linspace(0, 50, 51)
t = np.linspace(0, 10, 11)

mu = 0
sigma = 0.1
N = np.random.randn(n.shape[0]) * sigma + mu

y = np.sin(2*np.pi*(1/50)*n) + N
k = (2/(1*np.sqrt(2*np.pi))) * np.exp((-1/2) * ((t - 5)/2)**2) # mu = 5 e sigma = 2

y_conv_k = np.convolve(y, k)

plt.subplot(1,3,1)
plt.stem(n, y)
plt.subplot(1,3,2)
plt.stem(y_conv_k)
plt.subplot(1,3,3)
plt.stem(t, k)
```
