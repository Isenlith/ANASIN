## Exercício 1

$ x(t) = \text{cos}(t)  e^{0.3t}, \quad \{ \forall t \in \mathcal{R} : 0 \leq t \leq 10\} $

3. Pesquise e descreva o que é o efeito de *aliasing*, quando ele ocorre e maneiras de evitá-lo.

**Resposta**: O termo se refere a distorções entre o sinal original e o sinal final. Aliasing pode ocorrer com sinal sonoro (aliasing temporal) e visual (aliasing espacial). Essa distorção é o que gera um efeito de serrilhado.

Para evitá-lo é utilizada uma técnica de anti-aliasing, que se trata de um filtro. Ele é aplicado antes do processamento do sinal de modo a restringir a largura da banda deste sinal. Esse filtro não elimina por completo a distorção, ele resulta na divisão dos serrilhados em serrilhados menores, gerando o efeito de suavização.

Existem dois tipos de amostragem AA (antialiasing):

1. Aumentar a taxa de amostragem (FSAA, SSAA, MSAA, EQAA e CSAA)
2. Borrar os serrilhados (FXAA, MLAA, e SMAA), também chamado de post-AA ou AA de pós processamento)
