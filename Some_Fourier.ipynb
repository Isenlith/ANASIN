```
import numpy as np
from matplotlib import pyplot as plt
```

```
t = np.linspace(0, 1, 1000)

A = np.sin(2*np.pi*t*5)
B = np.sin(2*np.pi*t*7)
C = np.sin(2*np.pi*t*9)

mu = 0.5
sigma = 1
ruido = np.random.randn(t.shape[0]) * sigma + mu

sinal = 2*A + 7*B + 9*C

plt.plot(t, sinal)
```

```
#Transformada de Fourier
freq_i = 0
freq_f = 15

N = len(t)

X = []

for f in range(freq_i, freq_f):
  c = 2 * 1/N * np.abs(np.sum( sinal * np.exp(-2j*np.pi*f*t) ))
  X.append(c)
  
X = np.array(X)
plt.grid()
plt.stem(np.linspace(freq_i, freq_f, 15), X)
```
