```
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Audio
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [12, 5]
```

```
tempo = 1
framerate =  44200
frequencia = 322

t = np.linspace(0, tempo, framerate*tempo)
sinal = np.sin(2*np.pi*t*frequencia)

Audio(sinal, rate = framerate)
```

```
plt.plot(t, sinal)
```

```
re = np.sin(t*2*np.pi*264)
mi = np.sin(t*2*np.pi*330)

a = 0.4
b = 0.6

re_sus = a*re + b*mi

plt.plot(t, re_sus)
```

```
Audio(re_sus, rate = framerate)
```

```
tempo = 1
framerate =  44200
t = np.linspace(0, tempo, framerate*tempo)

re = np.sin(t*2*np.pi*264)
mi = np.sin(t*2*np.pi*330)

g = re + mi

N = len(t)
c = []

for k in range(0, N):
  c.append( (1/N) * g[k] * np.exp(-2j*np.pi*k*t[k]) ) 
  
c = np.real(c)

plt.plot(t, c)
```
